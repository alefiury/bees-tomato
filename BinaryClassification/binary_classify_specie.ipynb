```python
svm_param = [{'kernel': ['rbf'], 'gamma': [1e-2,1e-3,1e-4], 'C': [0.001, 0.1, 0.01, 1, 10], 'decision_function_shape': ['ovo']},
            {'kernel': ['poly'], 'gamma': [1e-2,1e-3,1e-4], 'C': [0.001, 0.1, 0.01, 1, 10], 'decision_function_shape': ['ovo']},
            {'kernel': ['sigmoid'], 'gamma': [1e-2,1e-3,1e-4], 'C': [0.001, 0.1, 0.01, 1, 10], 'decision_function_shape': ['ovo']}, 
            {'kernel': ['linear'], 'C': [0.001, 0.1, 0.01, 1, 10], 'decision_function_shape': ['ovo']}]

lr_param = [{'penalty': ['l1', 'l2']},
            {'C': [0.001, 0.1, 1, 10, 100]}]

rf_param = {'n_estimators': [100, 200],
            'max_features': ['auto', 'sqrt', 'log2']}

dtree_param = {"criterion": ["gini", "entropy"],
                "min_samples_split": [2, 10],
                "max_depth": [2, 5, 10]}

ens_param = {'lr__C': [0.001, 0.1, 1, 10, 100],
             'svc__C': [0.001, 0.1, 0.01, 1, 10]}
```

```python
def grid_search(X_train, y_train, model, param):   
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

    clf = GridSearchCV(model, param, scoring='f1_macro', cv=kfold)

    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    
    return clf.best_params_
```

```python
def metricas(X_test, y_test, clf):
    y_true, y_pred = y_test, clf.predict(X_test)
    
    print("Acurácia...: %.4f" %(metrics.accuracy_score(y_true, y_pred) * 100))
    print("Precision..: %.4f" %(metrics.precision_score(y_true, y_pred, average='macro') * 100))
    print("Recall.....: %.4f" %(metrics.recall_score(y_true, y_pred, average='macro') * 100))
    print("F1-Score...: %.4f" %(metrics.f1_score(y_true, y_pred, average='macro') * 100))
    print()
    print(metrics.classification_report(y_true, y_pred))
    print()
    print(pd.crosstab(y_true, y_pred, rownames=['True'], colnames=['Predicted'], margins=True))
    
    #metrics.plot_roc_curve(clf, X_test, y_test)  
    #plt.show()
    #fig = plt.figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
    #tab_acertos = sns.heatmap(pd.crosstab(y_true, y_pred, rownames=['True'], colnames=['Predicted'], margins=True), 
    #    cmap="YlGnBu", annot=True, annot_kws={'size':14}, cbar=False, square=True)
    #tab_acertos.set_xticklabels(tab_acertos.get_xticklabels(), rotation=45) 
    #tab_acertos.get_figure().savefig('heatmap.jpeg')
    #plt.show()
```

```python
def type_classification(df, buzz):
    if buzz == "flight":
        df = df[df['Annotation'] == 'voo']
    if buzz == "sonication":
        df = df[df['Annotation'] != 'voo']

    df = df.drop(['filename', 'Annotation'],axis=1)

    especies_list = df.iloc[:, -1]
    #encoder = LabelEncoder()
    #classes = encoder.fit_transform(especies_list)
    standard = StandardScaler()
    matrix = standard.fit_transform(np.array(df.iloc[:, :-1]))
    
    return matrix, especies_list
```

```python
matrix_voo, classes_voo   = type_classification(df, "flight")
matrix_son, classes_son   = type_classification(df, "sonication")
matrix_both, classes_both = type_classification(df, "both")
```

### Flight Classification

```python
X_train_voo, X_test_voo, y_train_voo, y_test_voo = train_test_split(matrix_voo, classes_voo, test_size=0.5, random_state=0)
```

```python
lr_modelo = LogisticRegression()
svm_modelo = SVC()
rf_modelo = RandomForestClassifier()
dtree_modelo = DecisionTreeClassifier()

svc = SVC()
rf = RandomForestClassifier()
lr = LogisticRegression()

modelos = [('svc', svc), ('rf', rf), ('lr', lr)]

ens_modelo = VotingClassifier(estimators=models, voting='hard')
```

```python
best_param_lr = grid_search(X_train_voo, y_train_voo, lr_modelo, lr_param)
best_param_svm = grid_search(X_train_voo, y_train_voo, svm_modelo, svm_param)
best_param_rf = grid_search(X_train_voo, y_train_voo, rf_modelo, rf_param)
best_param_dtree = grid_search(X_train_voo, y_train_voo, dtree_modelo, dtree_param)
best_param_ens = grid_search(X_train_voo, y_train_voo, ens_modelo, ens_param)
```

```python
def train_test(X_train, X_test, y_train, y_test, best_param_lr, best_param_svm, best_param_rf, best_param_dtree,
                best_param_ens):
    
    print("--------------------- Logistic Regression ---------------------")
    print()
    
    if list(best_param_lr.keys())[0] == 'C':
        lr_model = LogisticRegression(C=best_param_lr['C'])
    else:
        lr_model = LogisticRegression(penalty=best_param_lr['penalty'])

    lr_model.fit(X_train, y_train)

    y_pred = lr_model.predict(X_test)

    metricas(X_test, y_test, lr_model)
    print()
    
    print("--------------------- Support Vector Machine ---------------------")
    print()
    svm_model = SVC(C=best_param_svm['C'], decision_function_shape=best_param_svm['decision_function_shape'], 
                    gamma=best_param_svm['gamma'], kernel=best_param_svm['kernel'])

    svm_model.fit(X_train, y_train)

    y_pred = svm_model.predict(X_test)

    metricas(X_test, y_test, svm_model)
    print()
    
    print("--------------------- Random Forest ---------------------")
    print()
    rf_model = RandomForestClassifier(max_features=best_param_rf['max_features'], n_estimators=best_param_rf['n_estimators'])

    rf_model.fit(X_train, y_train)

    y_pred = rf_model.predict(X_test)
    
    metricas(X_test, y_test, rf_model)
    print()
    
    print("--------------------- Decision Tree ---------------------")
    print()
    dtree_model = DecisionTreeClassifier(criterion=best_param_dtree['criterion'], 
                                         max_depth=best_param_dtree['max_depth'], 
                                         min_samples_split=best_param_dtree['min_samples_split'])

    dtree_model.fit(X_train, y_train)

    y_pred = dtree_model.predict(X_test)

    metricas(X_test, y_test, dtree_model)
    print()
    
    print("--------------------- Ensemble ---------------------")
    print()
    svc = SVC(C=best_param_ens['svc__C'])
    rf = RandomForestClassifier()
    lr = LogisticRegression(C=best_param_ens['lr__C'])

    modelos = [('svc', svc), ('rf', rf), ('lr', lr)]

    ens_model = VotingClassifier(estimators=modelos, voting='hard')

    ens_model.fit(X_train, y_train)

    y_pred = ens_model.predict(X_test)

    metricas(X_test, y_test, ens_model)
    print()
    
    print("--------------------- Curva ROC ---------------------")
    print()
    fig, ax = plt.subplots()

    models = [
        ("LR", lr_model),
        ("SVM", svm_model),
        ("RF", rf_model),
        ("DTree", dtree_model),
        #("Ensemble", ens_model),
    ]

    model_displays = {}
    for name, pipeline in models:
        model_displays[name] = metrics.plot_roc_curve(pipeline, X_test, y_test, ax=ax, name=name)
    _ = ax.set_title('ROC curve')
```

```python
print("Results fligh classification")
print()

train_test(X_train_voo, X_test_voo, y_train_voo, y_test_voo, best_param_lr, best_param_svm, best_param_rf, best_param_dtree,
            best_param_ens)
```
