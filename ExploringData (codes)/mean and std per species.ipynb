```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Dados originais_artigo Priscila .xlsx")
```

```python
df.head()
```

```python
dados = df.drop(["Peso abelha (g)", "Tamanho tórax", "Duração (s)", "Média Freq..1",\
                            "Peso abelha (g).1", "Tamanho tórax.1", "Duração (s).1"], axis=1)
```

```python
dados.head()
```

```python
target_names = dados["Espécie - Abelha"].unique()
target_names
```

```python
lista_especies = []

for especie in target_names:
    #Filtra o DataFrame por espécie
    d = dados[dados["Espécie - Abelha"] == especie]
    
    #Junta todas as frequências por espécie
    #lista_freq = list(d[' Freq. 1 (Hz)']) + list(d[' Freq. 2 (Hz)']) + list(d['Freq. 3 (Hz)'])
       
    #Calcula o desvio padrão
    #dvp = np.std(lista_freq)
    dvp = np.std(d['Média Freq.'])
    
    #Calcula a média das médias
    media = d['Média Freq.'].mean()

    #Cria um dicionário com o desvio padrão e a média
    lista_especies.append([especie, media, dvp])
```

```python
media_desvio_padrao = pd.DataFrame(lista_especies, columns=['Espécies', 'Média', 'Desvio Padrão'])
```
